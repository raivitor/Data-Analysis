```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
url = "http://bigdatalabs.com.br/python/movie_metadata.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
df1 = df.dropna()
print(df1.isnull().sum())
print(df1.shape)
```

```python
print (df1.dtypes)
```

```python
df2 = df1.select_dtypes(include=['int64','float64']) #only float and int values
print (df2.shape)
```

```python
df2.describe()
```

```python
plt.figure(figsize = (25, 15))
sns.boxplot(x = 'title_year', y = 'imdb_score',  data = df1)
xt = plt.xticks(rotation=45)
```

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.distplot(df2["imdb_score"], kde=True, rug=False);
```

```python
corr = df2.corr()
f, ax = plt.subplots(figsize=(25, 20))
sns.heatmap(corr, linewidths=1,vmax=1.0, square=True, linecolor='black', annot=True)
```

```python
sns.lmplot(x="actor_1_facebook_likes", y="cast_total_facebook_likes", data=df1);
```
